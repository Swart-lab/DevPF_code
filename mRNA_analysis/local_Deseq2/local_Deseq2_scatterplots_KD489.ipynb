### Aim
This notebook creates volcano plots for the local Deseq2 differential expression analysis

```python
# import packages
import pandas as pd
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patheffects as pe
from functools import reduce

plt.rcParams["font.family"] = "Arial"
```

```python
def determine_significance(row):
    """
    set Log2FoldChange and adjusted p-value thresholds to filter for significant hits
    
    Keyword arguments:
    row -- pandas dataframe to filter on
    
    return values:
    significance level 
    """
    
    if row['log2FoldChange'] > 2 and row['padj'] < 0.01:
        return "significant_up"
    elif row['log2FoldChange'] < -2 and row['padj'] < 0.01:
        return "significant_down"
    else:
        return "not significant"
```

```python
def create_df(files):
    """
    read input files into pandas dataframe and add columns for -log10(adjusted p-value) and significance level
    
    Keyword arguments:
    files -- list of gzipped .csv files 
    
    return values:
    list with dataframes 
    """
    
    dfs = []
    for f in files:
        df = pd.read_csv(f, compression = 'gzip', sep=",")
        df['-log10(p)'] = np.log10(df['padj']) * (-1)
        df['significance'] = df.apply(determine_significance, axis=1)
        dfn = df[~ np.isnan(df['padj'])]
        dfs.append(dfn)
    return dfs
```

```python
def sig_count(df):
    """
    counts the number of significantly up- and downregulated hits in a pandas dataframe
    
    Keyword arguments:
    files -- pandas dataframe 
    
    return values:
    list with the number of significantly up and downregulated hits 
    """
    
    significant_up = (df['significance'] == 'significant_up').sum()
    significant_down = (df['significance'] == 'significant_down').sum()
    return [significant_up, significant_down]
```

```python
# note on file names:
# P = DevPF1; A = DevPF2; N = Nd7
# veg = onset (first cells in meiosis); 20 = early (20-40% fragmentatin);
# 80 = late (80-90% fragmentation); 100 = very late (6h after "late")

files = ['input_files/DE_results_resPNveg.csv.gz','input_files/DE_results_resPN20.csv.gz',
         'input_files/DE_results_resPN80.csv.gz','input_files/DE_results_resPN100.csv.gz',
        'input_files/DE_results_resANveg.csv.gz','input_files/DE_results_resAN20.csv.gz',
         'input_files/DE_results_resAN80.csv.gz','input_files/DE_results_resAN100.csv.gz',
        'input_files/DE_results_resAPveg.csv.gz','input_files/DE_results_resAP20.csv.gz',
         'input_files/DE_results_resAP80.csv.gz','input_files/DE_results_resAP100.csv.gz']

dfs = create_df(files)
```

```python
# check how many transcripts are available for analysis in each comparison
for df in dfs:
    print(len(df))
```

```python
# volcano plots for the knockdown comparisons at each time point

fig, axes = plt.subplots(3, 4,figsize = (16,9))
ax1, ax2, ax3, ax4 = axes[0]
ax5, ax6, ax7, ax8 = axes[1]
ax9, ax10, ax11, ax12 = axes[2]

ax = [ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8, ax9, ax10, ax11, ax12]

colors = ['violet', 'deepskyblue', 'white', 'grey']
category = ['significant_up', 'significant_down','not significant']
ax_count = int(-1)
x_max = []
x_min = []
y_max = []
num_sig = []

#groups = resPN100.groupby('significance')
for df in dfs:
    groups = df.groupby('significance')
    num_sig.append(sig_count(df))
    x_max.append(max(df['log2FoldChange']))
    x_min.append(min(df['log2FoldChange']))
    y_max.append(max(df['-log10(p)']))
    ax_count += 1
    c_count = int(-1)
    for cat in category:
        c_count += 1
        for name, group in groups:
            if name == cat:
                ax[ax_count].plot(group.log2FoldChange, group['-log10(p)'], marker='o', linestyle='', 
                                     markersize=5, label=name, markeredgecolor='black', 
                                     markerfacecolor = colors[c_count], markeredgewidth = 0.5)
ax_count = int(-1)
for a in ax:
    ax_count += 1
    a.set_xlim((x_min[ax_count] - 1), (x_max[ax_count] + 1))
    a.set_ylim(0, (y_max[ax_count] + 10))
    a.tick_params(axis='both', which='major', labelsize=15)
    a.axvline(x=0, ymin=0, ymax=7, ls='-', color='black', lw=1)
    a.annotate(f"{num_sig[ax_count][0]}", xy = (0.81, 0.9),xycoords='axes fraction',fontsize=17)
    a.annotate(f"{num_sig[ax_count][1]}", xy = (0.05, 0.9),xycoords='axes fraction',fontsize=17)
    
    if ax_count > 7:
        a.set_xlabel('log2(FoldChange)', wrap = True, fontsize = 17)
    if ax_count == 0:
        a.set_ylabel('DevPF1 vs ND7\n-log10(padj)', wrap = True, fontsize = 17)
    if ax_count == 4:
        a.set_ylabel('DevPF2 vs ND7\n-log10(padj)', wrap = True, fontsize = 17)
    if ax_count == 8:
        a.set_ylabel('DevPF2 vs DevPF1\n-log10(padj)', wrap = True, fontsize = 17)
    

#set titles
ax1.set_title('onset', wrap = True, fontsize = 22)
ax2.set_title('early', wrap = True, fontsize = 22)
ax3.set_title('late', wrap = True, fontsize = 22)
ax4.set_title('very late', wrap = True, fontsize = 22)


# label data poitns

ax4.legend(loc='center left', bbox_to_anchor=(1, 0.80), fancybox=True, fontsize = 15);
fig.tight_layout()

#fig.savefig(f"output_files/DevPF_DE_volcano_FDR01_P01_FC2.png", dpi = 300,\
#                    bbox_inches = 'tight')
```

```python
# store accession numbers for genes of interest in list
acc = pd.read_csv('input_files/Accessionnumbers.csv', sep=";", names=["Name", "PTET", "GSPAT", "GSPAT2",
                                                                      "GSPAT3"] )
acc = acc.drop(columns=["GSPAT", "GSPAT2","GSPAT3"])
acc = acc[~ pd.isna(acc['PTET'])]

interest = []

# extract the genes of interst from each of the dataframes
for df in dfs:
    #inter = df.loc[df['Unnamed: 0'].isin(interest_list)]
    inter = pd.merge(df, acc, left_on=["Unnamed: 0"], right_on=["PTET"], how="inner")
    inter = inter[['PTET', 'Name','log2FoldChange','padj','significance']]
    interest.append(inter)
    
```
